```python
datahandler = NewDataHandler()
datahandler.prepare_split(split = 'mental_health', seed = 42)
```

```python
TxGNNwNorm = TxGNNwNorm(data = datahandler, 
              weight_bias_track = False,
              proj_name = 'TxGNN',
              exp_name = 'TxGNN'
              )
```

```python
TxGNNwNorm.model_initialize(n_hid = 100, 
                      n_inp = 100, 
                      n_out = 100, 
                      proto = True,
                      proto_num = 3,
                      attention = False,
                      sim_measure = 'all_nodes_profile',
                      bert_measure = 'disease_name',
                      agg_measure = 'rarity',
                      num_walks = 200,
                      walk_mode = 'bit',
                      path_length = 2)
```

```python
TxGNNwNorm.pretrain(n_epoch = 2, 
               learning_rate = 1e-3,
               batch_size = 1024, 
               train_print_per_n = 20)
```

```python
TxGNNwNorm.finetune(n_epoch = 500, 
               learning_rate = 5e-4,
                      train_print_per_n = 5,
               valid_per_n = 20,
               save_name = finetune_result_path)
```
